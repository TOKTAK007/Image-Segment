การแปลงภาพสีเป็นภาพขาวดำเป็นการทำภาพเลขาคณิตเพื่อแปลงสีของแต่ละพิกเซลในภาพเป็นค่าสีเทา (grayscale) ซึ่งอยู่ในช่วงค่าเดียวกัน วิธีการทำนั้นจะใช้ค่าความสว่าง (luminance) ของสีในแต่ละพิกเซลเป็นพื้นฐาน
ใน PyTorch สามารถทำการแปลงภาพสีเป็นภาพขาวดำได้ด้วยฟังก์ชัน torchvision.transforms.ToTensor() และ torchvision.transforms.Grayscale() โดยใช้คำสั่งดังนี้

import torch

import torchvision.transforms as transforms

from PIL import Image

# โหลดภาพจากไฟล์
image_path = "image.jpg"
img = Image.open(image_path)

# แปลงภาพสีเป็น tensor
tensor_img = transforms.ToTensor()(img)

# แปลงภาพเป็นภาพขาวดำ
gray_img = transforms.Grayscale()(tensor_img)

# แปลงภาพขาวดำเป็น tensor
gray_tensor = transforms.ToTensor()(gray_img)

# แสดงผลภาพขาวดำ
print(gray_tensor.shape)

ในส่วนของ transforms.Grayscale() สามารถกำหนดจำนวนช่องสีที่ต้องการใช้ได้ โดยค่าเริ่มต้นคือ 1 ซึ่งหมายถึงภาพขาวดำที่ใช้เฉพาะช่องสีเท่านั้น ถ้าต้องการภาพขาวดำที่ใช้ทั้งช่องสี RGB สามารถกำหนดให้เป็น 3 ด้วย transforms.Grayscale(num_output_channels=3)

ใน tensor ที่เป็นภาพขาว-ดำ ค่าของแต่ละพิกเซลจะเป็นค่าเดียวกันในแต่ละช่อง ซึ่งหมายความว่า ถ้าค่าที่อยู่ใน Tensor มีค่ามากกว่าค่าที่กำหนดเป็น threshold จะถือว่าเป็นสีขาว และถ้าค่าน้อยกว่า threshold จะถือว่าเป็นสีดำ

ตัวอย่างเช่น หากเราต้องการแบ่งสีของ tensor img_tensor ที่มีขนาด (C, H, W) เป็นสีขาวและสีดำโดยกำหนด threshold เป็นค่า 128 จะทำได้โดยใช้โค้ดดังนี้

python


import torch

# สร้าง tensor ภาพขาว-ดำ แบบสุ่ม
img_tensor = torch.randint(low=0, high=256, size=(3, 256, 256), dtype=torch.uint8)

# กำหนด threshold เป็น 128
threshold = 128

# แบ่งสีขาว-ดำโดยใช้ threshold
white_tensor = (img_tensor > threshold).type(torch.uint8)

black_tensor = (img_tensor <= threshold).type(torch.uint8)

# ตรวจสอบขนาดของ tensor แต่ละตัว
print(white_tensor.shape)  # ผลลัพธ์ (3, 256, 256)

print(black_tensor.shape)  # ผลลัพธ์ (3, 256, 256)


โดยในตัวอย่างนี้ เรากำหนด threshold เป็น 128 และใช้ตัวดำเนินการเปรียบเทียบในการเช็คว่าแต่ละพิกเซลใน tensor img_tensor มีค่ามากกว่าหรือน้อยกว่า threshold โดยการใช้ > และ <= ตามลำดับ ซึ่งจะทำให้ได้ tensor white_tensor ที่มีค่าเป็น 1 ที่ตำแหน่งที่พิกเซลมีค่ามากกว่า threshold และค่าเหลือให้เป็น 0 และได้ tensor black_tensor ที่มีค่าเป็น 1 ที่ตำแหน่งที่พิกเซลมีค่าน้อยกว่าหรือเท่า

uint 8 เป็นชนิดของข้อมูล (data type) ที่ใช้เก็บค่าตัวเลขจำนวนเต็มบวกในภาษาโปรแกรมมิ่ง เช่น C, C++, Python เป็นต้น ซึ่ง uint 8 แทนค่า unsigned integer ที่ใช้เก็บข้อมูลที่มีค่าตั้งแต่ 0 ถึง 255 เท่านั้น โดยชนิดของข้อมูลนี้จะใช้พื้นที่หน่วยความจำ 1 byte เพื่อเก็บข้อมูลต่อหนึ่งค่า ซึ่งเป็นชนิดของข้อมูลที่ใช้บ่อยในการจัดการกับภาพดิจิตอล เนื่องจากค่าความเข้มของสีในภาพดิจิตอลมักจะมีค่าอยู่ในช่วง 0-255 ซึ่งเป็นช่วงของค่าที่ uint 8 สามารถเก็บได้

Gray tensor และ image tensor เป็นสองประเภทของ Tensor ที่ใช้ในการจัดเก็บข้อมูลภาพ โดยทั่วไป gray tensor จะเก็บข้อมูลภาพเป็นภาพขาวดำ (grayscale image) ซึ่งประกอบด้วยค่าความเข้มของสีเทาเฉด ๆ ต่าง ๆ ในภาพ ในขณะที่ image tensor จะเก็บข้อมูลภาพที่มีความหลากหลายของสี เช่น ภาพสี, ภาพโทนสี, ภาพ RGB, ภาพ RGBA หรือภาพอื่น ๆ ที่มีการใช้สีเป็นตัวแทนของข้อมูล

โดยปกติแล้ว gray tensor จะมี rank เท่ากับ 2 (2D tensor) ซึ่งเป็นขนาดของภาพขาว-ดำที่จัดเก็บไว้ ในขณะที่ image tensor อาจมี rank 2 ถึง 4 (2D ถึง 4D tensor) ซึ่งขนาดของ tensor จะแสดงถึงขนาดของภาพและจำนวนสีที่ใช้ในการแสดงผลภาพ ยกตัวอย่างเช่น ภาพสีที่แสดงผลด้วย RGB จะเก็บข้อมูลด้วย tensor ที่มี rank 3 (3D tensor) โดยมีขนาดของแต่ละ dimension คือ (ความสูง, ความกว้าง, จำนวนสี) โดยที่จำนวนสีมีค่าเท่ากับ 3 (สีแดง, สีเขียว และสีน้ำเงิน)

4D tensor เป็น Tensor ที่มี rank เท่ากับ 4 (4D tensor) ซึ่งประกอบด้วย dimension 4 ตัว โดยมีค่า index ในแต่ละ dimension เพื่ออ้างอิงถึงข้อมูลใน Tensor ดังนี้

torch.Size([2, 3, 128, 128])

ในตัวอย่างนี้ มี 4 dimension คือ

Dimension แรกมีค่าเท่ากับ 2 ซึ่งแทนจำนวนข้อมูลทั้งหมดใน Tensor นี้ (batch size)

Dimension ที่สองมีค่าเท่ากับ 3 ซึ่งแทนจำนวนช่องสี (channel) 
ที่ใช้ในการแสดงผลภาพ (RGB)

Dimension ที่สามมีค่าเท่ากับ 128 ซึ่งแทนความสูง (height) ของภาพ

Dimension ที่สี่มีค่าเท่ากับ 128 ซึ่งแทนความกว้าง (width) ของภาพ

ดังนั้น 4D tensor ในตัวอย่างนี้เป็น Tensor ที่มีขนาด 2x3x128x128 ซึ่งเก็บข้อมูลของภาพ 2 ภาพที่มีขนาด 128x128 พร้อมกับช่องสี 3 ช่องสี (RGB) ในแต่ละภาพ
