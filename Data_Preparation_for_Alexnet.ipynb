Threshold หมายถึง ระดับค่าที่กำหนดขึ้นมาเพื่อใช้ในการตัดสินใจหรือกรองข้อมูลตามเกณฑ์ที่กำหนดไว้ เช่น หากมีการกำหนด threshold ที่มีค่าเท่ากับ 50% ในการตัดสินใจว่าผลลัพธ์ของการทดสอบเป็น Positive หรือ Negative แล้วถ้าค่าความน่าจะเป็น (Probability) ของผลลัพธ์เกิน 50% จะถูกตัดสินใจว่า Positive และถ้าน้อยกว่า 50% จะถูกตัดสินใจว่า Negative ซึ่ง Threshold สามารถกำหนดได้ตามเหตุการณ์หรือความต้องการของการวิเคราะห์ข้อมูลในแต่ละครั้ง
